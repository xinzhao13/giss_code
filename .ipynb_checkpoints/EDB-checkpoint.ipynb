```python
import pandas as pd
import numpy as np
```

```python
DM=pd.read_excel('Book1.xlsx','DM')
DN=pd.read_excel('Book1.xlsx','DN')
P2b=pd.read_excel('Book1.xlsx','P2b')
P2bP=pd.read_excel('Book1.xlsx','P2bP')
```

```python
DM_O=DM.loc[DM['Origin Name'] == 'Edinburgh']
DM_D=DM.loc[DM['Dest Name'] == 'Edinburgh']
#####
DN_O=DN.loc[DN['Origin Name'] == 'Edinburgh']
DN_D=DN.loc[DN['Dest Name'] == 'Edinburgh']
#####
P2b_O=P2b.loc[P2b['Origin Name'] == 'Edinburgh']
P2b_D=P2b.loc[P2b['Dest Name'] == 'Edinburgh']
#####
P2bP_O=P2bP.loc[DM['Origin Name'] == 'Edinburgh']
P2bP_D=P2bP.loc[DM['Dest Name'] == 'Edinburgh']
```

```python
DN_D.head()
```

```python
Diff_DM_DN_D=pd.DataFrame(columns=['Origin','Dest','Origin Name','Dest Name','IVT_Diff','Wait_Diff','Boarding_Diff','GJT Lei_Diff','GJT Bus_Diff','GJT Com_Diff'])
Diff_DM_DN_O=pd.DataFrame(columns=['Origin','Dest','Origin Name','Dest Name','IVT_Diff','Wait_Diff','Boarding_Diff','GJT Lei_Diff','GJT Bus_Diff','GJT Com_Diff'])
```

```python
for i in range(0,4):
    Diff_DM_DN_D.iloc[:,i] = DM_D.iloc[:,i]
    Diff_DM_DN_O.iloc[:,i] = DM_O.iloc[:,i]
    
Diff_P2b_DM_D=Diff_DM_DN_D
Diff_P2bP_P2b_D=Diff_DM_DN_D
Diff_P2b_DM_O=Diff_DM_DN_O
Diff_P2bP_P2b_O=Diff_DM_DN_O
```

```python
for x in range(4,10):
    Diff_DM_DN_D.iloc[:,x] = DM_D.iloc[:,x] - DM_D['Origin'].map(DN_D.set_index('Origin').iloc[:,x-1])
    Diff_P2b_DM_D.iloc[:,x] = P2b_D.iloc[:,x] - P2b_D['Origin'].map(DM_D.set_index('Origin').iloc[:,x-1])
    Diff_P2bP_P2b_D.iloc[:,x] = P2bP_D.iloc[:,x] - P2bP_D['Origin'].map(P2b_D.set_index('Origin').iloc[:,x-1])
    Diff_DM_DN_O.iloc[:,x] = DM_O.iloc[:,x] - DM_O['Dest'].map(DN_O.set_index('Dest').iloc[:,x-1])
    Diff_P2b_DM_O.iloc[:,x] = P2b_O.iloc[:,x] - P2b_O['Dest'].map(DM_O.set_index('Dest').iloc[:,x-1])
    Diff_P2bP_P2b_O.iloc[:,x] = P2bP_O.iloc[:,x] - P2bP_O['Dest'].map(P2b_O.set_index('Dest').iloc[:,x-1])
```

```python
writer = pd.ExcelWriter('output.xlsx')
```

```python
    Diff_DM_DN_D.to_excel(writer,'Diff_DM_DN_D',index=False)
    Diff_P2b_DM_D.to_excel(writer,'Diff_P2b_DM_D', index=False)
    Diff_P2bP_P2b_D.to_excel(writer, 'Diff_P2bP_P2b_D',index=False)
    Diff_DM_DN_O.to_excel(writer,'Diff_DM_DN_O',index=False)
    Diff_P2b_DM_O.to_excel(writer,'Diff_P2b_DM_O',index=False)
    Diff_P2bP_P2b_O.to_excel(writer,'Diff_P2bP_P2b_O', index=False)
```
